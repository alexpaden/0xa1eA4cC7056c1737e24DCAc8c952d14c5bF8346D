```python
# CALL NEYNAR API WITH REQUESTS LIBRARY

import requests

url = "https://api.neynar.com/v1/farcaster/casts?fid=11528&limit=100"

headersX = {
    "accept": "application/json",
    "api_key": "NEYNAR API KEY HERE"
}

response = requests.get(url, headers=headersX)

print(response.text)
```

```python
# RESPONSE CLEANUP SECTION

texts = []

# Check if the response was successful
if response.status_code == 200:
    # Parse the response as JSON
    data = response.json()
    # Extract the list of casts from the response
    casts = data.get("result", {}).get("casts", [])
    # Iterate over the casts and append the text to the list
    for cast in casts:
        text = cast.get("text", "")  # Default to empty string if text is not found
        texts.append(text)
else:
    print(f"Failed to retrieve data: {response.status_code}")

# Output the array of texts
print(texts)
```

```python
# SECOND VERSION OF RESPONSE CLEANUP FOR RESPONSE ONLY, GENERATE A LIST OF CLEAN OPS
op_texts = []

# Check if the response was successful
if response.status_code == 200:
    # Parse the response as JSON
    data = response.json()
    # Extract the list of casts from the response
    casts = data.get("result", {}).get("casts", [])
    # Iterate over the casts and append the text to the list if both thread_hash and parent_hash are null
    for cast in casts:
        if cast.get("parentHash") is None:
            text = cast.get("text", "")  # Default to empty string if text is not found
            op_texts.append(text)
else:
    print(f"Failed to retrieve data: {response.status_code}")

# Output the array of texts
print(op_texts)
```

```python
formatted_texts = '\n'.join(op_texts)
print(formatted_texts)
```

```python
# GENERATE A SUMMARY OF THE MOST RECENT 3 OPS FROM YOUR ACCOUNT

from openai import OpenAI
OPENAI_API_KEY='OPENAPI KEY HERE'
client = OpenAI(api_key=OPENAI_API_KEY)

completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "Given these short thoughts create a summary of them all, and provide a list of classifications for each"},
    {"role": "user", "content": op_texts[0]},
    {"role": "user", "content": op_texts[1]},
    {"role": "user", "content": op_texts[2]}
  ]
)

print(completion.choices[0].message.content)
```

```python
formatted_texts = '\n'.join(op_texts)
#print(formatted_texts)

# GENERATE A SUMMARY OF THE MOST RECENT 3 OPS FROM YOUR ACCOUNT

completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "Given these short thoughts create a summary of them all, and provide a list of classifications for each"},
    {"role": "user", "content": formatted_texts}
  ]
)

print(completion.choices[0].message.content)
```

```python
# GENERATE a list of all messages from op posts as

# Generate user messages using a list comprehension
user_messages = [{"role": "user", "content": text} for text in op_texts]

# Complete list of messages including the system message
messages = [
    {"role": "system", "content": "Given these short thoughts create a summary of them all, and provide a list of classifications for each"}
] + user_messages

# Create the completion
completion = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=messages
)

# Print the completion result
print(completion.choices[0].message.content)
```
